```python
import json
def convert1(obj):
    l=[]
    obj=json.loads(obj)
    cnt=0
    for i in obj:
        if cnt <= 4:
            l.append(i['name']);cnt+=1
        else:
            break
    return l
```

# text matching on tags
# we will convert text onto vector

### various techniques used are : bag of words,tf,idf,word2vec
### we will use bag of words
### all tags are combined in one large text then extract that words that occur most times (through mapping) in large text 
### we consider 5000 words that occurs most time in text
### we remove stop words from most common words (eg.in,are,to...)

### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;w1 | w2 | w3 | w4 | w5 | ...... | w5000

### m1 &nbsp;&nbsp;&nbsp;5    | 3   |  0   |   ....................&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;| 1
### m2 &nbsp;&nbsp;&nbsp;2    | 0   |   1  |    .................   &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;| 3
### m3 &nbsp;&nbsp;&nbsp;
###   .
###   .
###   .
### m5000 

### (5000,5000) dimension

### Some example of stemming for root word "like" include:

### -> "likes"
### -> "liked"
### -> "likely"
### -> "liking"

# apply stemming 

```python
!pip install nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer() 
```

```python
print(ps.stem('danced'),ps.stem('loved'))
```

```python
def fun(text):
    y=set({})
    
    for i in text.split():
        y.add(ps.stem(i))
        
    y=list(y)
    
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(fun)
```

```python
!pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')

vectors = cv.fit_transform(movies['tags']).toarray()
# return scipy sparse matrix than convert into numpy array
```

# find cosine distance between every movies
# theta b/w two vector
# distance is inversely proportional to similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
# this is b/w 0 to 1
# near 1 high chances of similarity
# near 0 low chances of similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    movie_list=movie_list[1:5+1]
    
    for i in movie_list:
        print(movies.loc[i[0],'title'])
```
